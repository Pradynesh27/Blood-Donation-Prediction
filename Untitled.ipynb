```python
# Spliting the data and target variable
```

```python
X = Bld_df.drop(["Made Donation in March 2007"],axis = 1)
y = Bld_df["Made Donation in March 2007"]
```

```python
X
```

```python
y
```

```python
# Split taraining and testing data
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.1,random_state = 2)
```

```python
# Scale the features
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```python
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
```

```python
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
```

```python
#Trying Logistic Model
```

```python
# Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)
```

```python
y_pred = logreg.predict(X_test_scaled)
y_proba = logreg.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# Random Forest Mode
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# K nearest Neaighbor Model
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
y_pred = knn.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the KNN model: {accuracy * 100:.2f}%')
```

```python
# Support Vector Machine
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
# Cross Validation
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(kernel='linear')
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
results = {}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
    results[name] = (scores.mean(), scores.std())
    print(f"{name}: Mean Accuracy = {scores.mean():.4f}, Standard Deviation = {scores.std():.4f}")
```

### Hence Logistic regression achieves the heighest accuracy (76.04), suggesting it is the best performing model for this dataset  overall
